```python
from kubernetes import client
from kubernetes import config
from lib import utils
import json
```

```python
config.load_kube_config()
apiCustObject = client.CustomObjectsApi()
apiExtV1 = client.ApiextensionsV1Api() #  get api resources/custom resource definition
```

```python
crds = apiExtV1.list_custom_resource_definition()
```

```python
crds.items[0].spec
```

```python
group = crds.items[0].spec.group
plural = crds.items[0].spec.names.plural
version = None
for v in crds.items[0].spec.versions:
  print(f'version: {v.name}')
  version = v.name
```

```python
apiCustObject.list_namespaced_custom_object(group=group, version=version, namespace='default', plural=plural)
```

```python
namespace = ['kube-system']
for crd in crds.items:
  group = crd.spec.group
  plural = crd.spec.names.plural
  version = None
  for v in crd.spec.versions:
    version = v.name
  try:
    crd_data = apiCustObject.list_namespaced_custom_object(group=group, version=version, namespace=namespace[0], plural=plural)
    items = crd_data["items"]
    if len(items) > 0:
      print(f'version: {v.name}, group: {group}, plural: {plural}')
      print(f'{crd_data["kind"]}: {len(crd_data["items"])}')
  except:
    pass
```
